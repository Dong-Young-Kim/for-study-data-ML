### fillna() 함수

누락값을 원하는 값으로 채우는 함수이다.

````python
ebola.fillna(0)
````

모든 누락 값을 0으로 채울 수 있다.

````python
 ebola.fillna(method='ffill') # ffill = front fill
````

ffill 은 front fill로 앞에 있는 값으로 채우는 방식이다. 아직도 NaN 있는 field는 앞에 값이 존재하지 않기 때문이다.

````python
ebola.fillna(method='bfill') # ㅠfill = back fill
````

반대로 bfill은 뒤에 있는 값으로 채우는 함수이다.

### melt() 함수

melt 함수를 통해서 열에 중복 속성이 있는 경우, 열의 값을 행으로 melt 할 수 있다.
이때 고정하고 싶은 열은 'id_vars' 매개 변수에 list 형식으로 입력한다.

````python
pew2 = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')
pew2
````

모든 열을 하나의 행으로 만들었다. 첫 열인 'religion' 열은 고정되어 있어야 하므로 id_vars로 지정한다. 모든 열을 income 과 count 의 2 열로 변환한다.

위 데이터는 월 별로 최대 온도와 최저 온도만 기록되어 있으며 이외의 값은 모두 NaN 값으로 기록되어 있다.

````python
weather2 = pd.melt(weather, id_vars=['id','year','month','element'])
weather2
````

여기서 'tmax'와 'tmin'은 열의 속성으로 존재하는 것이 좋을 것 같아 보인다.

> 행에 있는 값을 열로 올려보자

### pivot_table() 함수

````python
weather3 = weather2.pivot_table(
    index = ['id', 'year', 'month', 'variable'],
    columns = 'element',
    values = 'value',
    dropna = False
)
weather3
````

tmax와 tmin이 새로운 열로 생겼다.
element 파라미터에는 새로운 columns가 될 값을 넣어주고 해당 열이 가져야 할 값은 values 파라미터로 입력한다.
dropna를 True 입력한다면 NaN 값이 있는 행을 삭제한다.

#### reset_index()
elememt 열도 위에 존재하여 원래의 tmax와 tmin이 element 열임을 나타낸다. 이를 reset_index() 함수를 통해서 없앨 수 있다.

### 중복된 행 처리하기 drop_duplicates()

중복된 행을 처리하는 방법에 대해서 알아보자

````python
billboard3 = billboard2[['year','artist','track','time']]
billboard3
````

billboard2에서는 wk와 score열의 값들로 인해서 중복된 데이터가 존재화지 않는다. 이 열들을 제거하고 생성한 billboard3의 데이터프레임은 중복된 행이 존재할 것이다.

````python
billboard3[billboard3['track']=='Bent']
````

위와 같이 track이 'Bent'인 값이 76개나 존재하는 것을 확인할 수 있다.

따라서 아래와 같이 중복된 행을 제거해보자

````python
billboard3.drop_duplicates()
````

위와 같이 drop_duplicates() 함수를 사용하면 알아서 중복된 행들을 삭제시켜 준다.
